```python
import requests
url=" https://api.football-data.org/v4/competitions/PL/matches?season=2024"
headers = {'X-Auth-Token':'60a77d091cba42969e15a63c8a8a3526'}

response = requests.get(url, headers=headers)
if response.status_code == 200:
    matches = response.json().get('matches', [])
    print(f"Number of matches: {len(matches)}") 
    for match in matches:
        print(f"{match['status']}, {match['homeTeam']['name']} vs {match['awayTeam']['name']}, Score: {match['score']['fullTime']['home']} - {match['score']['fullTime']['away']}")
else:      
    print(f"Failed to retrieve data: {response.status_code}")
    print(response.text)
    


```

```python
import pandas as pd
match_data=[]
for match in matches:
    match_data.append({
        'status':match['status'],
        'homeTeam':match['homeTeam']['name'],
        'awayTeam':match['awayTeam']['name'],
        'homeScore':match['score']['fullTime']['home'],
        'awayScore' :match['score']['fullTime']['away']
    })

data=pd.DataFrame(match_data)
data
    
    
```

Show the win probablity for all premier league teams in the 2024-2025 season .
 

```python
# Counting all matches
total_matches = len(match_data) 
print(f"Total matches: {total_matches}")


# Showing the win probability for each premier league teams in the 2024-2025 season.
#initializing a dictionary to store win probabilities
win_probabilities = {}

#calculating the number of wins and total games played
for team in data['homeTeam'].unique():
    # home wins
    home_wins = len(data[(data['homeTeam'] == team) & (data['homeScore'] > data['awayScore'])])
    #away wins
    away_wins = len(data[(data['awayTeam'] == team) & (data['awayScore'] > data['homeScore'])])
    # Total wins for the team
    total_wins = home_wins + away_wins
    # Total games played by the team
    total_games = len(data[(data['homeTeam'] == team) | (data['awayTeam'] == team)])
    #win probability
    if total_games > 0:
        win_probabilities[team] = total_wins / total_games
    else:
        win_probabilities[team] = 0.0
# Converting the win probabilities dictionary to a DataFrame and sort it(ASCENDING)
win_probabilities_df = pd.DataFrame(list(win_probabilities.items()), columns=['Team', 'Win Probability'])
win_probabilities_df = win_probabilities_df.sort_values(by='Win Probability', ascending=False)  
# Displaying the win probabilities DataFrame
win_probabilities_df
```
